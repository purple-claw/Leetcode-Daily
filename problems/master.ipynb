#### Febuary 21 - Special Binary String

```python
# Special Binary String
def solfeb21(s):
    # if the length of the string is less than 2 chars Simply return
    if not s:
        return ""
    blcks = []
    blc = 0
    sIdx = 0
    for i,char in enumerate(s):
        blc += 1 if s[i] == '1' else -1 # we need to check the top level string, if it exists then we will proceed
        
        # check if the counter is empty or not
        if blc == 0:
            sbstr = s[sIdx + 1:i]
            opSbstr = solfeb21(sbstr)
            blcks.append("1" + opSbstr + "0")
            sIdx += 1
    blcks.sort(reverse=True)
    return "".join(blcks)

s  = "11011000"
print(solfeb21(s))
```

### Febuary 22 - (762) Prime Number of Set Bits in Binary Representation

```python
def solfeb22(l,r):
    prms = {2, 3, 5, 7, 11, 13, 17, 19} # For the Binaru num range l and r, we have only take the prime nums from 2 to 19 only.
    cnt = 0 # counter to count the values
    for i in range(l,r+1):
        sbin = bin(i)[2:].count('1')
        if sbin in prms:
            cnt += 1
    return cnt

print(solfeb22(l=6, r=10))
```

### Febuary 26 - Reducing the Number in Binary Representation

#### Given the binary representation of an integer as a string s, return the number of steps to reduce it to 1 under the following rules:

If the current number is even, you have to divide it by 2.

If the current number is odd, you have to add 1 to it.

It is guaranteed that you can always reach one for all test cases.

 

Example 1:

Input: s = "1101"
Output: 6
Explanation: "1101" corressponds to number 13 in their decimal representation.
Step 1) 13 is odd, add 1 and obtain 14. 
Step 2) 14 is even, divide by 2 and obtain 7.
Step 3) 7 is odd, add 1 and obtain 8.
Step 4) 8 is even, divide by 2 and obtain 4.  
Step 5) 4 is even, divide by 2 and obtain 2. 
Step 6) 2 is even, divide by 2 and obtain 1.  
Example 2:

Input: s = "10"
Output: 1
Explanation: "10" corresponds to number 2 in their decimal representation.
Step 1) 2 is even, divide by 2 and obtain 1.  
Example 3:

Input: s = "1"
Output: 0
 

Constraints:

1 <= s.length <= 500
s consists of characters '0' or '1'
s[0] == '1'

```python
def solfeb26(s):
    cnt = 0
    dec = int(s,2)
    if dec == 1:
        return 0

    return recur(dec,cnt)

def recur(n, cnt):
    if n == 1:
        return cnt
    cnt += 1
    if n % 2 == 0:
        n //= 2
    else:
        n += 1
    return recur(n,cnt)

print(solfeb26("1101"))
```
